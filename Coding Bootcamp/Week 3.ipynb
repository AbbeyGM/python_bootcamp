# Elif Statement - else if statement

```python
# using the elif conditional statement
x, y = 5, 10
if x > y:
    print("x is greater")
elif x < y:
    print("x is less")
```

```python
# checking more than one elif conditional statement
x, y = 5, 10
if x > y:
    print("x is greater")
elif (x + 10) < y: # checking if 15 is less than 10 - we wrap the maths in a () so it gets executed first
    print("x is less")
elif (x + 5) == y: # checking if 10 is equal to 10
    print("equal")
```

```python
# writing mltiple conditionals within each other - multiple block levels
x, y, z = 5,10,5
if x>y:
    print("x is greater than y")
elif x<= y:
    if x==z:
        print("x is equal to z") # resulting output
    elif x != z:
        print("x is not equal to z") #won't get a hit bc above is correct
```

All elif statements are connected to one original if statement, so that once a single conditional is True, the rest do not run. Let’s see an example:

```python
# testing output of two if statements in a row that are both true
x, y, z = 5, 10, 5
if x < y:
    print("x is less")
if x==z:
    print("x is equal")
```

```python
# testing output of an if and elif statement that are both true
x, y, z = 5, 10, 5
if x < y:
    print("x is less")
elif x == z:
    print("x is equal to z")     # doesn't print bc the if statement is true - so doesn't go to the elif
```

# Friday Project - Creating a Calculator

```python
# step 1: ask user for calculation to be performed
operation = input("Would you like to add/subtract/multiply/divide? ").lower( )
print( "You chose {}.".format(operation) ) # for testing purposes
```

```python
# step 2: ask for numbers, alert order matters for subtracting and dividing
if operation == "subtract" or operation == "divide":
    print("You chose {}".format(operation))
    print("Please keep in mind that the order of your numbers matter.")
num1 = input("What is your first number?")
num2 = input("What is the second number?")
print("First number: {}".format(num1))     #For testing purposes
print("Second number: {}".format(num2))
```

```python
# step 3: setup try/except for mathematical operation
try:
    #immediately try to convert numbers into floats
    num1, num2 = float(num1), float(num2)
    if operation == "add":
        result = num1 + num2
        print("{} + {} = {}".format(num1,num2,result))
    elif operation == "subtract":
        result = num1 - num2
        print( "{} - {} = {}".format(num1, num2, result) )
    elif operation == "multiply":
        result = num1 * num2
        print( "{} * {} = {}".format(num1, num2, result) )
    elif operation == "divide":
        result = num1 / num2
        print( "{} / {} = {}".format(num1, num2, result) )
    else:
    # else will be hit if they didn't chose an option correctly
        print("Sorry, but '{ }' is not an option.".format(operation) )
except:
# Print error
    print("Error: Improper numbers used. Please try again")
```
