```python
import numpy as np
import matplotlib.pyplot as plt

#creating a list using list comprehension
x = [x for x in range(10)]

print(x)

#a manual for loop to create the same ist, this one has more lines of cod and is leff effecient
x1 = []
for i in range(10):
    x1.append(i)

print(x1)

#use list comprehension again to create our y values
y = [x**2 for x in x]       #squaring values of in the list x
print(y)


#plot
plt.figure(figsize=(8,6))
plt.scatter(x,y, color='red')

#add labels and titles
plt.title("Quadratic Function", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)

#Display the plot
plt.grid(True)
plt.savefig("Quadratic.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,1000)  #note np linspace(start,stop,numer_of_points) - creating linear array using numpy
print(x)

y = [x**2+1 for x in x]
print(y)

# Plot


#plot
plt.figure(figsize=(8,6))
plt.plot(x,y, color='blue',linewidth=1) #now does smooth line connecting instead of scatter

#add labels and titles
plt.title("Quadratic Function", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)

#Display the plot
plt.grid(True)
plt.savefig("Quadratic_continuous.png")
plt.show()

```
